```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2
from geopandas.tools import geocode

host = '172.17.0.2'
port = '5432'
dbname = 'postgres'
username = 'postgres'
pwd = 'kpass'

conn = psycopg2.connect("host='{}' port={} dbname='{}' user={} password={}".format(host, port, dbname, username, pwd))

# radius in decimal degree
r = (1/111320) * 300 # change 300 to what you want. The unit is meter.

# input address that you want to find
addr= '23 Lewis Ave, Brooklyn, NY' # example

lon = geocode(addr, provider='nominatim').geometry[0].centroid.x
lat = geocode(addr, provider='nominatim').geometry[0].centroid.y

sql = "select * \
        from final \
        where ST_Intersects(ST_GeomFromText(final.the_geom), ST_Buffer( \
            ST_GeomFromText('Point({} {})'),{}));".format(lon, lat, r)

# sql = "select * from final"

dat = sqlio.read_sql_query(sql, conn)
# conn = None

dat.nb = dat.nb.astype(float)

import geopandas as gpd
from shapely import wkt
tmp = sqlio.read_sql_query("select * \
        from nyct2010 \
        where ST_Intersects(ST_GeomFromText(nyct2010.the_geom), ST_Buffer( \
            ST_GeomFromText('Point({} {})'),{}));".format(lon, lat, r)
                          , conn)
tmp['the_geom'] = tmp['the_geom'].apply(wkt.loads)
tmp = gpd.GeoDataFrame(tmp, geometry='the_geom')
tmp.crs = {'init': 'epsg:4326', 'no_defs': True}

# 1 mile = 1609.34 meter
import folium

m = folium.Map(location=[float(lat), float(lon)], zoom_start=15, tiles='cartodbpositron')
folium.Choropleth(geo_data=tmp.the_geom, fill_color='Red', fill_opacity=0.7).add_to(m)
for i in [300, 500, 700]:
    folium.Circle([float(lat), float(lon)], i, fill=True).add_to(m)
m
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('ggplot')
```

## Ethnic group

```python
ethnic = ['B01001A_001E', 'B01001B_001E','B01001D_001E']
```

```python
tmps = []
for year in years:
    tmp = []
    for e in ethnic:
        tmp.append(sum(dat[dat.year==year].nb[(dat.vid == e)]))
    tmps.append(tmp)
```

```python
f, ax = plt.subplots(1, 3,figsize=(18,5))
for i,year in zip(range(0,3),years):
    labels = 'White', 'Black', 'Asian'
    sizes = tmps[i]
    colors = ['gold', 'yellowgreen', 'lightcoral']#, 'lightskyblue']

    # Plot
    ax[i].pie(sizes, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)

    ax[i].axis('equal')
    ax[i].set_title(year, fontsize=24)
```
